```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import re
```

```python
# load the datasets
fake = pd.read_csv("Fake.csv")
true = pd.read_csv("True.csv")
```

```python
# understand the data
print(fake.shape)
print(true.shape)
```

```python
display(fake.head(3))
display(true.head(3))
```

```python
# concatinate fake and true
fake["label"] = 0
true["label"] = 1

df = pd.concat([fake, true], ignore_index=True)

print(df.shape)
display(df.head(3))
```

```python
# i think date may not help the model so i dropped it
df = df.drop(["date"], axis = 1)
```

```python
# also drop the subject
df = df.drop(["subject"], axis = 1)
```

### Test Preprocessing

```python
print(df["content"].iloc[0])
```

```python
# combine title and text 
df["content"] = df["title"] + " " + df["text"]
```

```python
# convert into lower case
df["content"] = df["content"].str.lower()
```

```python
#remove puntuation & special characters
def clean_text(text): 
    text = re.sub(r"[^a-zA-Z\s]", '', text)
    return text

df["content"] = df["content"].apply(clean_text)
```

```python
print(df["content"].iloc[0])
```

```python
# remove stop words
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words = set(stopwords.words("english"))

def remove_stop_words(text):
    return " ".join(word for word in text.split() if word not in stop_words)

df["content"] = df["content"].apply(remove_stop_words)
```

```python
print(df["content"].iloc[0])
```

```python
# tokenization
df["tokens"] = df["content"].apply(lambda x: x.split())
```

```python
print(df["tokens"].iloc[0])
```

```python
import nltk
nltk.download('wordnet')
```

```python
# Lemmartization (converts words to their dictionary form)
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
lemmatizer = WordNetLemmatizer()

df["tokens"] = df["tokens"].apply(lambda words: [lemmatizer.lemmatize(word) for word in words])
```

```python
print(df["tokens"].iloc[0])
```

```python
# join tokens back
df["processed_text"] = df["tokens"].apply(lambda x: " ".join(x))
```

```python
df.head(3)
```

```python
# split the dataset
from sklearn.model_selection import train_test_split

train_val_df, test_df = train_test_split(
    df, 
    test_size=0.2,
    random_state=42,
    stratify=df["label"]
)

train_df, val_df = train_test_split(
    train_val_df,
    test_size=0.2,
    random_state=42,
    stratify=train_val_df["label"]
)
```

```python
# only select processed_text and label
train_df = train_df[["processed_text", "label"]]
val_df   = val_df[["processed_text", "label"]]
test_df  = test_df[["processed_text", "label"]]
```

```python
# save them into seperate csv files
train_df.to_csv("train.csv", index=False)
val_df.to_csv("val.csv", index=False)
test_df.to_csv("test.csv", index=False)
```
